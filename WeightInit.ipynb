```python
import torch
import torch.nn as nn
```

```python
layer = nn.Linear(5, 5)
layer.weight
```

```python
nn.init.uniform_(layer.weight, a = 0, b = 3)
```

```python
print(nn.init.normal_(layer.weight, mean = 0, std = 1))
print('\n')
print(nn.init.normal_(layer.weight, mean = 0, std = 0.2))
```

```python
nn.init.constant_(layer.bias, 0) # sets all biases to zero
print(layer.bias)

print('\n')

nn.init.zeros_(layer.bias) # sets all biases to zero
print(layer.bias)
```

```python
print(nn.init.xavier_uniform_(layer.weight, gain = 1))
print('\n')
print(nn.init.xavier_normal_(layer.weight, gain = 1))
```
