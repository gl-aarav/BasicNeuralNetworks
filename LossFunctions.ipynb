```python
import torch
import torch.nn as nn
```

```python
prediction = torch.randn(4, 5)
label = torch.randn(4, 5)
```

```python
print(prediction)
print('\n')
print(label)
```

```python
mse = nn.MSELoss(reduction = 'none')
loss = mse(prediction, label)
loss
```

```python
mse = nn.MSELoss(reduction = 'sum')
loss = mse(prediction, label)
loss
```

```python
mse = nn.MSELoss(reduction = 'mean')
loss = mse(prediction, label)
loss
```

```python
import numpy as np
x = prediction.numpy()
y = label.numpy()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
print(x)
print('\n')
print(sigmoid(x))
```

```python
x = sigmoid(x)
```

```python
loss_values = []
for i in range(len(y)):
    batch_loss = []
    for j in range(len(y[0])):
        if y[i][j] == 1:
            loss = -np.log(x[i][j])
        else:
            loss = -np.log(1 - x[i][j])
        batch_loss.append(loss)
        
    loss_values.append(batch_loss)

print(np.mean(loss_values))
```
